```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import differential_evolution
```

```python
def fun(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

# Task2: Global optimization

```python
dots = np.arange(1, 30+0.0001, 0.001)
```

```python
plt.figure(figsize = (10,6))
plt.plot(dots, fun(dots))
plt.legend(['function'])
plt.show()
```

```python
res = differential_evolution(fun, [(1,30)])
answ = res['fun'][0]
res
# передается список кортежей - границ оптимизации, сколько кортежей - столько переменных у функции
```

```python
with open('submission_2.txt', 'w') as file:
    file.write('{:.2f}'.format(answ))
    
```
