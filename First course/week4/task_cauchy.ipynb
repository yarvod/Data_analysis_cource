## Разберемся с распределением Коши

Функция распределения (нормальная, параметры $x_0 = 0, \; \gamma = 1$):
$$F(x) = \frac{1}{\pi} \arctan{\left( \frac{x - x_0}{\gamma} \right)} + 0.5$$
Плотность вероятности:
$$f(x) = \frac{1}{\pi \gamma \left[ 1 + \left(\frac{x-x_0}{\gamma} \right)^2 \right]}$$

Отметим, что данное распределние не имеет мат ожидания, медиана равна $x_0$, мода равна $x_0$, дисперсия $+\infty$.

### 

```python
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import cauchy
```

```python
def cauchy_dens(x, x_0 = 0, gamma = 1):
    return 1/(np.pi * gamma * (1 + ((x-x_0)/gamma)**2))

def cauchy_dist(x,x_0=0,gamma = 1):
    return 1/np.pi * np.arctan((x-x_0)/gamma) + 1/2
```

```python
dots = np.random.uniform(0, 10, (1, 100))[0]
dot = np.arange(-5,5,0.01)
```

```python
fig, ax = plt.subplots(figsize = (10,6))
n, bins, patches = ax.hist(cauchy.rvs(size = 1000),1000, density = True, histtype='bar')
ax.plot(bins, cauchy_dens(bins), '--')
plt.xlim(-25,25)


ax.set_title('Koshi distribution')
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')
plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize = (18,6))

ax[0].plot(dot,cauchy_dens(dot), '--')
ax[1].plot(dot, cauchy_dist(dot), '--')
ax[0].set_title('Cauchy density')
ax[0].set_xlabel('$x$')
ax[0].set_ylabel('$f(x)$')
ax[1].set_title('Cauchy distribution')
ax[1].set_xlabel('$x$')
ax[1].set_ylabel('$f(x)$')
plt.show()
```
