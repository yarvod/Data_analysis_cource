```python
import numpy as np
import re
import pandas as pd
import scipy
from scipy.spatial.distance import cosine
import copy
```

```python
# запишем текст из файла в строку и приведем к нижнему регистру
text = ''
with open('sentences.txt') as infile:
    for line in infile:
        text += line.lower()
```

```python
# разобьем стоку на слова с помощью регулярки
text_splitted = re.split('[^a-z]', text)
```

```python
# первый способ убрать все пропуски
for i in range(text_splitted.count('')):
    text_splitted.remove('')
```

```python
# второй способ убрать все пропуски
while 1:
    try:
        text_splitted.remove('')
    except ValueError:
        print('not "" here right now! ')
        break
```

```python
# сделаем массив из строк из всех предложений
sentences = re.split('[\n]', text)
while 1:
    try:
        sentences.remove('')
    except ValueError:
        print('not "" here right now! ')
        break
```

```python
# все слова без повторений
allwords = sorted(list(set(text_splitted)))
```

```python
# запишем в матрицу количество вхождений данного слова в данное предложение
a = np.zeros((len(sentences),len(allwords)))
for i in range(len(sentences)):
    for j in range(len(allwords)):
#         a[i][j] = len(re.findall(r'\b{}\b'.format(allwords[j]), sentences[i]))
        count = 0
        for k in range(len(re.split(r'[^a-z]', sentences[i]))):
            if re.split(r'[^a-z]', sentences[i])[k] == allwords[j]:
                count += 1

        a[i][j] = count
        
        
pd.DataFrame(a, columns = allwords)
```

```python
# найдем косинусное расстояние между 1м предложением и остальными
cos = []
for i in range(1,len(sentences)):
    cos.append(cosine(a[i], a[0]))
cos1 = copy.copy(cos)
#первый минимум
min_1 = min(cos1)
print('min_1 = ',min_1)
# второй минимум
cos1[cos1.index(min_1)] = max(cos)
min_2 = min(cos1)
print('min_2 = ',min_2)
```

```python
ind1 = cos.index(min_1) + 1
ind2 = cos.index(min_2) + 1
print('ind_1 = ',ind1)
print('ind_2 = ',ind2)
```

```python
with open('submission-1.txt', 'w') as file:
    file.write('{} '.format(ind1)+'{}'.format(ind2))
    
```
