```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.linalg import solve
```

```python
def fun(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
plt.figure(figsize = (10,6))
dots = np.arange(1,15.01,0.1)
plt.plot(np.arange(0,16,0.1), fun(np.arange(0,16,0.1)))
```

## 1й порядок

```python
b = np.array([fun(1),fun(15)])[:,np.newaxis]
A = np.array([[1,1],[15,1]])
sol1 = solve(A,b).reshape(1,2)[0]
def pol1(x, a,b):
    return a*x+b
plt.figure(figsize = (10,6))
plt.plot(dots, fun(dots))
plt.plot(dots, pol1(dots,*sol1), 'r', ls = '--')
```

## 2й порядок

```python
b = np.array([fun(1),fun(8), fun(15)])[:,np.newaxis]
A = np.array([[1,1,1],[64,8,1],[225,15,1]])
sol2 = solve(A,b).reshape(1,3)[0]
def pol2(x, a,b,c):
    return a*x**2+b*x + c
plt.figure(figsize = (10,6))
plt.plot(dots, fun(dots))
plt.plot(dots, pol2(dots,*sol2), 'r', ls = '--')
```

## 3й порядок

```python
b = np.array([fun(1),fun(4),fun(10), fun(15)])[:,np.newaxis]
A = np.array([[1,1,1,1],[64,16,4,1],[1000,100,10,1],[3375,225,15,1]])
sol3 = solve(A,b).reshape(1,4)[0]
def pol3(x, a,b,c,d):
    return a*x**3+b*x**2 + c*x + d
plt.figure(figsize = (10,6))
plt.plot(dots, fun(dots))
plt.plot(dots, pol3(dots,*sol3), 'r', ls = '--')
```
